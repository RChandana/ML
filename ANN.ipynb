```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/Churn_Modelling.csv')
```

```python
data
```

```python
data.head(10)
```

```python
x = data.drop(['RowNumber','Age','Surname','Geography','Gender'], axis =1)
```

```python
x.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xi = scaler.fit_transform(x)
```

```python
y = data['Balance']
```

```python
x_train,x_test,y_train,y_test = train_test_split(xi,y,test_size=0.2,random_state=0)
```

```python
reg = MLPRegressor(hidden_layer_sizes=(64,64,64),activation='relu',alpha=0.0001, batch_size='auto',learning_rate='constant',learning_rate_init=0.001,random_state=1,max_iter=2000).fit(x_train,y_train)
```

```python
j=reg.predict(x_test)
```

```python
plt.scatter(y_test,j)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
mean_squared_error(y_test, j)
```

```python
r2_score(y_test,j)
```

```python
mean_absolute_error(y_test,j)
```
